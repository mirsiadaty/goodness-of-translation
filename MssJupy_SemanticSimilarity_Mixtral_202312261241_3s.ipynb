```python

## we expect the LLM scoring the same sentence pair consistently across the replicates
# also:
## select one per each replicate set




# MssJupy_SemanticSimilarity_Mixtral_202312261241.ipynb : Score entire 1.4K sents: en-pers-en.
ARitexts = ArArDSMBGSents # for purpose of this cell only!






#
print('ReplicationCountFD' ,ReplicationCountFD)
#
print(len(ARitexts))
print('ARitexts' ,ARitexts)



#ArArDSMBGSentsNonReplicated=[]
#
#for cntr0,decoded221 in enumerate( ArArDSMBGSents ):
NonReplicatedIndex = list( range(0,len(ArArDSMBGSents),3) )
#
print('NonReplicatedIndex' ,len(NonReplicatedIndex))
print('NonReplicatedIndex' ,NonReplicatedIndex)




import re




# MssJupy_SemanticSimilarity_Mixtral_202312261241.ipynb : Score entire 1.4K sents: en-pers-en.
#
ArScores101 =[]
ArrLlmRspnses33txtNonRepl =[]
ArNonhomogenRplcts =[]
#
for cntr1,valu1 in enumerate(NonReplicatedIndex):
    #
    #score = ArArDSMBGSents[NonReplicatedIndex[LinearCounter]]
    score = ArArDSMBGSents[valu1]
    #
    score = re.sub(r'%',r'', score)
    score = int(score)
    # make this llm 0 distinct from lower-left triangle 0's
    #if (score==0) :
    #    score=score+1
    #
    print(cntr1,valu1,score)
    ArScores101.append(score)
    #
    ArrLlmRspnses33txtNonRepl.append(ArrLlmRspnses33txtFD[valu1])
    
    # we expect the LLM scoring the same sentence pair consistently across the replicates
    # HARDCODED! ASSUMEs ReplicationCount ==3! !!!!!!!
    Rplct2 = ArArDSMBGSents[valu1+1]
    Rplct3 = ArArDSMBGSents[valu1+2]
    #
    Rplct2 = re.sub(r'%',r'', Rplct2)
    Rplct2 = int(Rplct2)
    Rplct3 = re.sub(r'%',r'', Rplct3)
    Rplct3 = int(Rplct3)
    #
    if (score==Rplct2 and score==Rplct3) :
        # good!
        x=2
    else :
        ArNonhomogenRplcts.append([cntr1,valu1 , score,Rplct2,Rplct3])
    
#
# QA
print('ArNonhomogenRplcts' ,ArNonhomogenRplcts)
assert len(ArNonhomogenRplcts) == 0





# MssJupy_SemanticSimilarity_Mixtral_202312251407.ipynb: 1.use iset18 (more anagramic sents), 2.uniform heatmap format.
# use heat color set where lower left triangle is -20, BUT THEN the heat color range is based on min and max of observed scores.
MinArScores101 = min(ArScores101)
MaxArScores101 = max(ArScores101)
print('MinArScores101' ,MinArScores101)
print('MaxArScores101' ,MaxArScores101)




print('99' ,time.asctime( time.localtime( time.time() ) ))
```
